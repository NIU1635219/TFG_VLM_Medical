# Dual validation data export

Este notebook genera dos paquetes de validacion, uno por modelo:
- qwen3_5-9b@q8_0
- google/gemma-4-e4b@q8_0

Cada paquete escribe solo la data web en:
- `clinical_eval_app/data/App_Autonoma/<EXECUTION_ID>/`
- `clinical_eval_app/data/App_Asistida/<EXECUTION_ID>/`

Dentro de cada ejecucion solo se generan:
- `cases.json`
- `cases_export_debug.csv`
- `images/`

La seleccion se hace cruzando los escenarios A y F para cada modelo por separado, de modo que ambos escenarios queden representados con ambos modelos.
El identificador de ejecucion sigue el formato `YYYYMMDD_HHMMSSF`, como en el notebook 04.
Este flujo no copia la app completa; solo exporta `cases.json`, `cases_export_debug.csv` e imagenes compartidas.

Se usan los ultimos runs q8 detectados por modelo:
- qwen: A en `20260505_132318` y F en `20260423_171622`

```python
from __future__ import annotations

import importlib
import json
import sys
from pathlib import Path

base = Path.cwd().resolve()
project_root = next((candidate for candidate in [base, *base.parents] if (candidate / 'pyproject.toml').exists()), base)
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

import src.preprocessing.prepare_dual_validation as prepare_dual_validation
prepare_dual_validation = importlib.reload(prepare_dual_validation)

from src.preprocessing.prepare_dual_validation import DualValidationConfig, build_execution_id, export_dual_validation_web_data

print(f'PROJECT_ROOT: {project_root}')

SCENARIO_A_NAME = 'scenario_A'
SCENARIO_F_NAME = 'scenario_F'
QWEN_MODEL_ID = 'qwen3_5-9b@q8_0'
GEMMA_MODEL_ID = 'google/gemma-4-e4b@q8_0'
QWEN_RUN_ID_A = '20260505_132318'
QWEN_RUN_ID_F = '20260423_171622'
GEMMA_RUN_ID_A = '20260506_150112'
GEMMA_RUN_ID_F = '20260429_141632'
RUN_PATH_OVERRIDE_A = None
RUN_PATH_OVERRIDE_F = None
TOP_K = 50
NORMALIZE_BY_CLASS = True
NORMALIZATION_MODE = 'proportional'
VALID_CLASSES = ('AD', 'HP', 'ASS')
EXECUTION_ID = build_execution_id(suffix='D')

WEB_DATA_ROOT = project_root / 'clinical_eval_app' / 'data'
SOURCE_APP_DIR = project_root / 'clinical_eval_app'

shared_config = {
    'scenario_a_name': SCENARIO_A_NAME,
    'scenario_f_name': SCENARIO_F_NAME,
    'top_k': int(TOP_K),
    'normalize_by_class': bool(NORMALIZE_BY_CLASS),
    'normalization_mode': NORMALIZATION_MODE,
    'valid_classes': tuple(VALID_CLASSES),
    'output_dist_dir': WEB_DATA_ROOT,
    'source_app_dir': SOURCE_APP_DIR,
    'app_a_name': 'App_Autonoma',
    'app_f_name': 'App_Asistida',
}

qwen_config = {
    'qwen_model_id': QWEN_MODEL_ID,
    'qwen_run_id_a': QWEN_RUN_ID_A,
    'qwen_run_id_f': QWEN_RUN_ID_F,
    'qwen_run_path_override_a': RUN_PATH_OVERRIDE_A,
    'qwen_run_path_override_f': RUN_PATH_OVERRIDE_F,
}

gemma_config = {
    'gemma_model_id': GEMMA_MODEL_ID,
    'gemma_run_id_a': GEMMA_RUN_ID_A,
    'gemma_run_id_f': GEMMA_RUN_ID_F,
    'gemma_run_path_override_a': RUN_PATH_OVERRIDE_A,
    'gemma_run_path_override_f': RUN_PATH_OVERRIDE_F,
}

dual_config = DualValidationConfig(**shared_config, **qwen_config, **gemma_config)

paths, selection, export_result = export_dual_validation_web_data(
    config=dual_config,
    execution_id=EXECUTION_ID,
    start=project_root,
)

summary = {
    'execution_id': EXECUTION_ID,
    'selected_image_ids': export_result.selected_image_ids,
    'selected_count': len(export_result.selected_image_ids),
    'app_a_cases_json': str(export_result.output_app_a_cases_json),
    'app_f_cases_json': str(export_result.output_app_f_cases_json),
    'app_a_export_csv': str(export_result.output_app_a_export_csv),
    'app_f_export_csv': str(export_result.output_app_f_export_csv),
    'app_a_images_dir': str(export_result.output_app_a_images_dir),
    'app_f_images_dir': str(export_result.output_app_f_images_dir),
    'dist_selected': selection.dist_selected,
    'dist_candidates': selection.dist_candidates,
    'skipped_by_reason': selection.skipped_by_reason,
}

print(json.dumps(summary, ensure_ascii=False, indent=2))
```
